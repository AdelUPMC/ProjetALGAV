```python

#on represente le bit de poids fort à gauche
#version naïve
#les clés codés sur 128 bits sont représentées dans des listes de 128 elements

#renvoi la position du bit de poids fort de key
def pospf(cle):
    taille=len(cle)
    for i in range(taille):
        if(cle[i]==1):
            return i
    #cas où la clé vaut 0
    return taille
    
def inf(cle1,cle2):
    #determine si cle1<cle2
    #cas 1: position du bit de poids fort de key1 différent de celui de key2
    p1=pospf(cle1)
    p2=pospf(cle2)
    #print("pospf cle1:"+str(p1))
    #print("pospf cle2:"+str(p2))
    
    if(p1<p2):
        return False
    elif (p1>p2):
        return True
    
    #cas où les bits de poids fort de key1 et key2 sont
    #à la même position :comparaison bit à bit entre key1 et key2,
    #on sort de la boucle si on a trouvé 2 bits différent
    else:
        for i in range(p1,len(cle1)):
            if(cle1[i]>cle2[i]):
                return False
            elif(cle1[i]<cle2[i]):
                return True
        #bits sont égaux
        return False
    
def eg(cle1,cle2):
    p1=pospf(cle1)
    p2=pospf(cle2)
    if(p1!=p2):
        return False
    else:
        for i in range(p1,len(cle1)):
            if(cle1[i]!=cle2[i]):
                return False
        return True



cle1=[0,0,1,1,1,1,0,1]
cle2=[0,0,0,0,1,1,0,1]
#cle4=[0,1,1,0]

#eg(cle3,cle4)
#inf(cle1,cle2)
#pospf(cle3)
        

        
        
```
